## top 10 most populated countries

```python
#filter out total population
total_pop= df[df['Series Code']=='SP.POP.TOTL']

#sort the total pop of year 2022
total_pop_sorted= total_pop.sort_values(by="2022",ascending=False)

#extract top 10 countries from the sorted pop
total_pop_topten= total_pop_sorted.head(10)

print(total_pop_topten['Country Code'])
```

## Bar Plot

```python
#creating a bar plot for year 2022
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sb.barplot(x="2022", y="Country Code", data=total_pop_topten, palette="coolwarm")
plt.title("Top 10 Most Populated Countries in 2022", fontsize=10)
plt.xlabel("Total population", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.show

#creating a bar plot for year 2015
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sb.barplot(x="2015", y="Country Code", data=total_pop_topten, palette="coolwarm")
plt.title("Top 10 Most Populated Countries in 2015", fontsize=10)
plt.xlabel("Total population", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.show
```

## Extracting bottom 10 wrt. total population

```python
#sort the total pop of year 2022
total_pop_sorted1= total_pop.sort_values(by="2022",ascending=True)

#extract top 10 countries from the sorted pop
total_pop_bottomten= total_pop_sorted1.head(10)

print(total_pop_bottomten['Country Code'])
```

## Extracting top 10 countries wrt. highest male & female population resp.

```python
#filter out total male  population
male_pop= df[df['Series Code']=='SP.POP.TOTL.MA.IN']

#sort the total male pop of year 2022
male_pop_sorted= male_pop.sort_values(by="2022",ascending=False)

#extract top 10 countries from the sorted male pop
male_pop_topten= male_pop_sorted.head(10)

print(male_pop_topten['Country Code'])
```

```python
#filter out total female population
female_pop= df[df['Series Code']=='SP.POP.TOTL.FE.IN']

#sort the total female pop of year 2022
female_pop_sorted= female_pop.sort_values(by="2022",ascending=False)

#extract top 10 countries from the sorted female pop
female_pop_topten= female_pop_sorted.head(10)

print(female_pop_topten['Country Code'])
```

```python
#creating a bar plot for male_pop in 2022
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sb.barplot(x="2022", y="Country Code", data=male_pop_topten, palette="viridis")
plt.title("Top 10 Male Populated Countries in 2022", fontsize=10)
plt.xlabel("Total  Male population", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.show

#creating a bar plot for female_pop in 2022
plt.figure(figsize=(15,6))
plt.subplot(2,2,1)
sb.barplot(x="2022", y="Country Code", data=female_pop_topten, palette="viridis")
plt.title("Top 10 Female Populated Countries in 2022", fontsize=10)
plt.xlabel("Total Female population", fontsize=10)
plt.ylabel("Country", fontsize=10)
plt.show
```
